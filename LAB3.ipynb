This lab focuses on applying Exploratory Data Analysis (EDA) techniques to explore and understand the dataset. Various visualizations were used to identify patterns, relationships, and key factors that may influence students' academic performance.

```python
import pandas as pd
df = pd.read_csv("Student_performance_data _.csv")
df.head()
```

```python
df.describe(include='all')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['GPA'], bins=20)
plt.title("Distribution of GPA")
plt.show()
```

The histogram shows that GPA values are fairly spread across the range, with most students clustered around the middle values. Very low and very high GPAs are less common.

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='StudyTimeWeekly', y='GPA', data=df)

plt.title("Study Time vs GPA")
plt.xlabel("Study Time Weekly")
plt.ylabel("GPA")
plt.show()
```

The scatter plot shows a positive relationship between weekly study time and GPA. Students who spend more time studying tend to achieve higher GPAs.

```python
plt.figure(figsize=(8,5))

sns.barplot(x='ParentalSupport', y='GPA', data=df)

plt.title("Parental Support vs GPA")
plt.xlabel("Parental Support")
plt.ylabel("Average GPA")

plt.show()
```

The bar chart shows that students who receive higher parental support tend to achieve better GPA compared to those with less support.

```python
df.groupby('Gender')['GPA'].mean()
```

```python
df2 = df.drop("StudentID", axis=1)
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(df2.corr(),
            annot=True,
            fmt=".2f",
            cmap="coolwarm",
            linewidths=0.5)

plt.title("Correlation Heatmap")
plt.show()
```

The heatmap shows the relationships between different variables. GPA has a positive correlation with study time and parental support, while absences show a negative correlation with GPA. This indicates that students who study more and receive more support tend to have higher academic performance.
